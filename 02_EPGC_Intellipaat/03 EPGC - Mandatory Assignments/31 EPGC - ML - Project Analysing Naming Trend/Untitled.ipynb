```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import io
from zipfile import ZipFile
import re
```

```python
zip_path = 'names.zip'
```

```python
with ZipFile(zip_path, "r") as z:
    pattern = re.compile(r"^yob\d{4}\.txt$")
    files = [f for f in z.namelist() if pattern.match(f)]

    frames = []
    for f in files:
        dfy = pd.read_csv(io.BytesIO(z.read(f)),encoding='utf-8',header=None,names=['name','sex','births'])
        dfy['year'] = int(f[3:7])
        frames.append(dfy)
    df = pd.concat(frames,ignore_index=True)
```
