### 3 & 4. `Stemming` & `Lemmatization`

- Stemming using porter

```python
from nltk.stem import PorterStemmer
import time
```

```python
words = ["running", "flies", "easily", "fairly", "happiness", "arguing", "automation", "studies", "organization", "studying", "everyone", "transforming",
         "stunning", "better", "antagonizing", "good", "goods", "goodies", "liking", "likes",  "hurry","hurried","begin", "beginning", "begins", "unwell", "unhappy", "best"]
```

```python
stemmer = PorterStemmer()
```

```python
### Start Timing
start_time = time.time()


### Run stemming
stemmed = [stemmer.stem(word) for word in words]

end_time = time.time()
time_elapsed = (end_time - start_time)
```

```python
print("Original Words:", words,"\n")
print("Stemmed Words:", stemmed, "\n")
print(f"Time taken for stemming: {time_elapsed : .3f} sec") 
```

- Lemmatizer using WordNet

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
### Download the required NLTK resources
nltk.download('wordnet')
nltk.download('omw-1.4') #optional but helps with the better lemmatization
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
### Start Timing
start_time = time.time()

### Run Lemmatization
lemmatized = [lemmatizer.lemmatize(word) for word in words]

end_time = time.time()
time_elapsed = (end_time - start_time)
```

```python
print("Original Words:", words,"\n")
print("Lemmatized Words:", lemmatized, "\n")
print(f"Time taken for lemmatization: {time_elapsed : .3f} sec") 
```
